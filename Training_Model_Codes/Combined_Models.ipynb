# Neural Network Model

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
```

```python
df = pd.read_csv('../Resources/data.csv')
df = df.drop(['id', 'Unnamed: 32'], axis=1)
df
```

```python
null_check = df.isnull().sum()
null_check
```

```python
final_df = pd.get_dummies(df, columns=['diagnosis'], dtype=int)
final_df
```

```python
# Splits data into features and target arrays
y=final_df['diagnosis_M']
X=final_df.drop(['diagnosis_M'], axis=1)

# Splits the data into training & testing datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train.columns)
hidden_nodes_layer1 =  8
hidden_nodes_layer2 = 5

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu"))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled,y_train,epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

# SVM Model - All Columns

```python
import pandas as pd
import numpy as np
import os
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report
import joblib
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("../Resources/data.csv")

df = df.drop(['Unnamed: 32', 'id'], axis=1)
df
```

```python
final_df = pd.get_dummies(df, columns=['diagnosis'], dtype=int)
final_df
```

```python
target = final_df["diagnosis_M"]
target_names = ["1", "0"]
```

```python
features_data = final_df.drop(["diagnosis_B", "diagnosis_M"], axis=1)
feature_names = features_data.columns
features_data.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(features_data, target, random_state=1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
model = SVC(kernel='linear')
model.fit(X_train_scaled, y_train)
```

```python
predictions = model.predict(X_test_scaled)
print(classification_report(y_test, predictions,
                            target_names=target_names))
```
